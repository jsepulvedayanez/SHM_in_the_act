```r
# Instalar y cargar la librería
if (!require("beepr")) {
  install.packages("beepr")
}
library(beepr)

notificar_fin <- function(mensaje = "Proceso terminado", repeticiones = 10, tiempo_entre_beeps = 0.5) {
  for (i in 1:repeticiones) {
    beep(2)  # Sonido tipo 2 (puedes cambiarlo del 1 al 11)
    Sys.sleep(tiempo_entre_beeps)  # Retraso entre cada beep
  }
  cat("\n¡", mensaje, "!", sep="")
  cat("\nHora de finalización:", format(Sys.time(), "%H:%M:%S"), "\n")
}

# Ejemplo de uso:
# Tu proceso aquí...
# Al final solo agregas:

```

```r
notificar_fin("Mi proceso terminó")
```
